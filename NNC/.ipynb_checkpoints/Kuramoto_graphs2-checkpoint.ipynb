```python
# -----------------------------------------------------------
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(4, 3))

# Define X-axis positions (data indices: 0, 1, 2, ...)
x_positions = np.arange(len(hnc_means))

# Define X-axis labels (desired layer counts: 1, 2, 3, ...)
x_labels = x_positions + 1


plt.plot(

    x_positions,

    hnc_means,

    '-',

    label=r"$\text{HNC}_{t,\mu}$",

    color='steelblue',

    linewidth=2,

    alpha=0.8

)

plt.fill_between(

    x_positions,

    hnc_means - hnc_stes,

    hnc_means + hnc_stes,

    color='steelblue',

    alpha=0.2

)


plt.plot(

    x_positions,

    thnc_means,

    '-',

    label=r"$\text{HNC}_{t}$",

    color='orange',

    linewidth=2,

    alpha=0.8

)

plt.fill_between(

    x_positions,

    thnc_means - thnc_stes,

    thnc_means + thnc_stes,

    color='orange',

    alpha=0.2

)

# Plot 2: ENC Data with Error Bars
plt.plot(

    x_positions,

    enc_means5,

    '-',

    label="ENC",

    color='black',

    linewidth=2,

    alpha=0.8

)

plt.fill_between(

    x_positions,

    enc_means5 - enc_stes5,

    enc_means5 + enc_stes5,

    color='black',

    alpha=0.2

)

# ADDED HORIZONTAL LINE FOR OC VALUE

oc_value = 3.649243

oc_err = 0.01566105  # Replace with your actual error


# Your x-axis limits (should match your plot's x-range)

x_min, x_max = plt.xlim()  # Or set manually, e.g., x_min=0, x_max=10


plt.axhline(

    y=oc_value, 

    color='blue', 

    linewidth=1.5, 

    linestyle='--',

    label='vAGM'

)


# Add error band

plt.fill_between(

    [x_min, x_max], 

    oc_value - oc_err, 

    oc_value + oc_err, 

    color='blue', 

    alpha=0.2, 


)


# --- X-AXIS MODIFICATION START ---
# Set the tick positions and the new labels
plt.xticks(x_positions, x_labels)
# --- X-AXIS MODIFICATION END ---
plt.xlim([0,4])

# --- Formatting ---
plt.xlabel("Number of hidden layers of a (head) hypernetwork", fontsize=10)
plt.ylabel("Loss", fontsize=10)
plt.title("Test loss vs number of hidden layers", fontsize=10)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.legend(loc='center left', bbox_to_anchor=(0.65, 0.42),fontsize=9)
#plt.yscale('log')#
plt.tight_layout()
plt.savefig("loss_vs_number_of_layers_kuramoto.pdf", bbox_inches='tight', pad_inches=0.2)
print('loss_vs_layers_modified_x_ticks.png')
```

```python
from matplotlib.lines import Line2D
plt.figure(figsize=(4, 3))

# --- main curves ---
plt.plot(hnc_params, hnc_means, '-',
         label=r"$\mathrm{HNC}_{t,\mu}$",
         color='steelblue', linewidth=1.0, alpha=0.8)

plt.plot(thnc_params, thnc_means, '-',
         label=r"$\mathrm{HNC}_{t}$",
         color='tab:orange', linewidth=1.0, alpha=0.8)

# --- ENC curves (no legend labels here) ---
l2, = plt.plot(enc_params2, enc_means2, '-', color='silver',  linewidth=1.0, alpha=0.8)
l3, = plt.plot(enc_params3, enc_means3, '-', color='dimgrey', linewidth=1.0, alpha=0.8)
l4, = plt.plot(enc_params4, enc_means4, '-', color='black',   linewidth=1.0, alpha=0.8)

# --- custom legend ---
enc_header = Line2D([], [], linestyle='none', label=r"ENC:")
enc_2 = Line2D([], [], color='silver',  label="2 L")
enc_3 = Line2D([], [], color='dimgrey', label="3 L")
enc_4 = Line2D([], [], color='black',   label="4 L")

handles = [
    plt.gca().lines[0],  # HNC_{t,}
    plt.gca().lines[1],  # HNC_t
    enc_header,
    enc_2,
    enc_3,
    enc_4,
]

plt.legend(handles=handles,fontsize=9)

plt.xlabel("Number of parameters")
plt.ylabel("Mean Loss")
plt.title("Loss vs total number of parameters")
plt.tight_layout()
plt.savefig("loss_vs_number_of_parameters_kuramoto.pdf", bbox_inches='tight', pad_inches=0.2)
plt.tight_layout()
plt.show()
```
