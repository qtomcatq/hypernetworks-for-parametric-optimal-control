```python
dim=10
N=dim

#A,B=generate_cogwheel_matrices(dim)
A=np.array([[1,0],[1,0]])
B=np.array([[1],[0]])
x00=np.array([1,0.5])
%store A
%store B
%store x00
```

```python
nsteps=64
total=1
dt=total/nsteps
N=np.shape(x00)[0]
Ac = A*dt + np.eye(A.shape[0])
Bc = B*dt

#print (Ac,Bc)

Qfc = np.eye(N)
Q = 0 *Qfc
Rc = 0.000000001
R = Rc*np.eye(B.shape[1])*dt
print(A)
print(B)
print(x00)
```

```python
contenergy = []
toterrors = []
totenergy = []
inferrors = []

ste_contenergy=[]
ste_toterrors=[]
ste_totenergy=[]

x0=x00
rng=10
desired_state_xf=0*x00
multipliers = np.linspace(0.75, 1.25, 40)
print(multipliers)
matrices = [m * A for m in multipliers]
for j in tqdm(range(0,40)):
    iter = 1
    desired_state_xf=np.tile(0*x00,(iter,1))
    actual_state_x, x1 = np.tile(x00,(iter,1)), np.tile(x00, (iter,1))
   
        
    gamma=0 
    Ac=matrices[j]*dt+np.eye(A.shape[0])
    optimal_control_input = lqr(actual_state_x, 
                                        desired_state_xf, 
                                        Q, R, Ac, Bc, nsteps, Qfc)
  
    cycle_tenergy=[]
    cycle_errors =[]
    cycle_cenergy=[]
    cycle_inferrs=[]
    
    
    


    energy=0
    inferr=0
    
    
    for l in (range(nsteps)):
  
        state_error = x1 - desired_state_xf                
        inf_error=actual_state_x-x1
        inf = np.linalg.norm(inf_error, axis=1)
         
        action=np.einsum("ij, kj -> ki", optimal_control_input[l], state_error) 
      
    
        actual_state_x =  np.einsum("ij, kj -> ki", Ac, actual_state_x) + np.einsum("ij, kj -> ki", Bc, action)
     
        x1 = actual_state_x
            

        energy+=  np.mean(np.einsum("kj,kj->k", action,np.einsum("ij, kj -> ki", R/Rc, action)))
        
                  
        inferr+=dt*np.mean(inf**2)             
        real_error=actual_state_x - desired_state_xf
        sem = np.mean(np.sum(real_error**2,axis=1))

        ener=np.squeeze(np.asarray(energy)) 

    contenergy.append(energy) 
    toterrors.append(sem)
    totenergy.append(energy+sem)  
    inferrors.append(inferr)

    print(energy)
    print(sem)
    print(energy+sem)
    print(inferr)
        
        
        
       

    
t = np.linspace(0, rng, rng)    
print("control energy")    
print(contenergy)
print("distance")
print(toterrors)
print("tot energy")
print(totenergy)
%store totenergy
%store contenergy
%store toterrors
```
