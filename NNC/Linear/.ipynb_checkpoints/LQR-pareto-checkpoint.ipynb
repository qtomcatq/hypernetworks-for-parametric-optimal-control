```python
import numpy as np
from tqdm import tqdm

contenergy = []
toterrors = []
totenergy = []
inferrors = []

ste_contenergy = []
ste_toterrors = []
ste_totenergy = []

x0 = x00
rng = 40
desired_state_xf = 0 * x00
multipliers = np.linspace(1, rng, rng)
print(multipliers)

for j in tqdm(range(0, rng)):
    iter = 1
    desired_state_xf = np.tile(0 * x00, (iter, 1))
    actual_state_x, x1 = np.tile(x00, (iter, 1)), np.tile(x00, (iter, 1))
    
    gamma = 0 
    Ac = A * dt + np.eye(A.shape[0])
    R = Rc * R_base * 10 ** (multipliers[j] / 4)
    print(R)
    optimal_control_input = lqr(actual_state_x, 
                                desired_state_xf, 
                                Q, R, Ac, Bc, nsteps, Qfc)
    Ac= A *dt
    cycle_tenergy = []
    cycle_errors = []
    cycle_cenergy = []
    cycle_inferrs = []
    
    energy = 0
    inferr = 0
    
    for l in range(nsteps):
        state_error = x1 - desired_state_xf                
    
        action = np.einsum("ij, kj -> ki", optimal_control_input[l], state_error) 
        
        # Heun's method for state update
        # Step 1: Compute intermediate state (Euler step)
        k1 = np.einsum("ij, kj -> ki", Ac, actual_state_x) + np.einsum("ij, kj -> ki", Bc, action)
        
        # Step 2: Compute slope at intermediate state
        intermediate_state = actual_state_x +  k1
        if l<(nsteps-1):
     
            intermediate_action = np.einsum("ij, kj -> ki", optimal_control_input[l+1], intermediate_state-desired_state_xf)
        else:
            intermediate_action = np.einsum("ij, kj -> ki", optimal_control_input[l], intermediate_state-desired_state_xf)
        
        k2 = np.einsum("ij, kj -> ki", Ac, intermediate_state) + np.einsum("ij, kj -> ki", Bc, intermediate_action)
        
        # Step 3: Average slopes and update state
        actual_state_x = actual_state_x + (1 / 2.0) * (k1 + k2)
        
        x1 = actual_state_x
            
        energy += np.mean(np.einsum("kj,kj->k", action, np.einsum("ij, kj -> ki", R_base, action)))
        
        real_error = actual_state_x - desired_state_xf
        sem = np.mean(np.sum(real_error**2, axis=1))
        
        ener = np.squeeze(np.asarray(energy))  

    contenergy.append(energy) 
    toterrors.append(sem)
    totenergy.append(energy + sem)  

    print(energy)
    print(sem)
    print(energy + sem)
 
t = np.linspace(0, rng, rng)    
print("control energy")    
print(contenergy)
print("distance")
print(toterrors)
print("tot energy")
print(totenergy)
%store totenergy
%store contenergy
%store toterrors
```
