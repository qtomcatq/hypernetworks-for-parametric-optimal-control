```python
import numpy as np
import pdb

import control
import matplotlib.pyplot as plt
import torch
import torchsde
from torch import nn, profiler, Tensor, nan, vmap
import copy
import matplotlib.pyplot as plt

from tqdm import tqdm
import numpy as np
import pdb

import os
import sys

project_root = os.path.abspath(os.path.join(os.getcwd(), '../..')) 
sys.path.append(project_root)  # Add folder_a to sys.path
sys.path.append(os.path.join(project_root, '../..', 'Networks'))  # Add folder_b to sys.path

from Networks.utils import get_gpu_memory,objective, adj_lattice, criticalK,lqr,Kgain, generate_system,iterr, normal, ema_update,mass_append, update_ema, generate_system_canon, load_clean_state_dict, sample_linear
from helpers import   HyperCoeffsLinearControlStochasticLQRImpl,  find_params
import psutil

from IPython.display import clear_output
from adabelief_pytorch import AdaBelief

is_cuda = torch.cuda.is_available()
device = 'cuda' if is_cuda else 'cpu'

if not is_cuda:
    print("Warning: CUDA not available; falling back to CPU but this is likely to be very slow.")
import gc
path=os.path.abspath(os.getcwd())
gc.collect()
torch.cuda.empty_cache()
```

```python
dim=128

batch_size=8*256

weight=0.1

A,B=generate_system_canon(dim)



```

```python
learning_rate=0.001
alpha=0.999
gamma=0.9999
period=5000
t_total=1
iter=20000
state_size=dim
global_perf=[]
global_eperf=[]
global_stperf=[]
global_grad=[]
global_steperf=[]
global_eener=[]
global_steener=[]
global_resutls=[]

for layers in range(1,5):
    seed=0
    
    #torch.manual_seed(seed)
    At=torch.tensor(A).float().to(device) 
    Bt=torch.tensor(B).float().to(device) 
    coremat=(At.unsqueeze(0)).repeat(batch_size,1,1)
    
    
    diff=0
    
    strategy="HNC"
    SML= HyperCoeffsLinearControlStochasticLQRImpl(At,Bt,batch_size, device,layers,strategy).to(device) 
    
    params =  find_params(SML)
    
    controller_optimizer = AdaBelief(params, lr=learning_rate, eps=1e-16, betas=(0.9,0.999), weight_decouple = True, rectify = True, amsgrad=False)
    scheduler = torch.optim.lr_scheduler.ExponentialLR(controller_optimizer, gamma=gamma)
    
    
    SML.batch=batch_size
    
    t = np.linspace(0, iter, iter)
    
    t_size=int(32)
              
    score= 10000000
    
    ema_model = copy.deepcopy(SML)
    
                
    ema_loss=0
    ema_ener=0
    ema_std=0
    enr_std=0
    
    ema_test_loss=0
    ema_test_ener=0
    ema_test_std=0
    enr_test_std=0
    
    ts_exp = torch.linspace(0, t_total, t_size).float().to(device)
    
    eperf=[]
    stperf=[]
    eloss=[]
    stloss=[]
    emaperf=[]
    emastperf=[]
    emaener=[]
    emastener=[]
    gradnorms=[]
    results=[]
   
    
    for j in tqdm(range(iter)):

        amat, x0= sample_linear(dim,batch_size, device)
       
        ini_enr = torch.zeros(batch_size, 1).to(device)
    
        ys_exp = torch.cat((x0,ini_enr),dim=1).to(device)
        
    
     
        coremat[:,-1,:]=amat
        SML.At= coremat
        SML.Bt= Bt
        SML.poly.amat=amat


        ys_tray = torchsde.sdeint_adjoint(SML, ys_exp, ts_exp,method='reversible_heun', dt=ts_exp[1]-ts_exp[0],
                             adjoint_method='adjoint_reversible_heun',)
        
                
        print(psutil.virtual_memory())
        dif=torch.sum(ys_tray[-1,:,0:state_size]**2,dim=1)
        

       
        dif1=weight*ys_tray[-1,:,-1]   
        diff=dif+dif1
    
        lossvec= torch.mean(diff) 
        logtrick= torch.mean(torch.log(diff))
        stdvec=torch.std(diff)
    
        
        print('total loss')
        print(lossvec)
        energy= torch.mean(dif1)
        stdener= torch.std(dif1)
        
        
        logtrick.backward()
        
        total_sq_norm=0    
        for p in SML.parameters():
            if p.grad is not None:
                param_norm = p.grad.detach().data.norm(2)
                total_sq_norm += param_norm.item() ** 2
        print("grad norm")
        print(total_sq_norm)
        gradnorms.append(total_sq_norm)
        results.append(diff.cpu().detach().numpy())
        controller_optimizer.step()
        controller_optimizer.zero_grad()
        scheduler.step()
 
        update_ema(SML, ema_model, alpha)
        [ema_loss, ema_ener, enr_std, ema_std]=ema_update(j,alpha, lossvec,energy,stdener,stdvec,ema_loss,ema_ener,enr_std,ema_std)
       
        
        print('ema loss')
        print(ema_loss)
        print('ema_ener')
        print(ema_ener)
        
        [eperf, stperf, eloss, stloss, emaperf, emaener,emastperf,emastener]= mass_append(lossvec,stdvec,energy,stdener, batch_size, ema_loss, ema_ener, enr_std, ema_std, eperf, stperf, eloss, stloss, emaperf, emaener,emastperf,emastener)
       
    
        if score>np.sqrt(ema_loss):
            score=np.sqrt(ema_loss)
    
            totener=ema_loss
            ener=ema_ener           
            errs=totener-ener
            error_vector=ema_std**2
    
            #torch.save(ema_model.state_dict(), path+"/" + strategy + "ema" + str(layers)+ ".pth")
            #torch.save(SML.state_dict(), path+"/" + strategy  + str(layers)+ ".pth")
            print('model saved')
        if lossvec>500:
            break
        
        if j%200==0:
            gc.collect()
            clear_output(wait=True)
        del lossvec, stdvec 
    global_stperf.append(stperf)    
    global_perf.append(eperf)
    global_eperf.append(emaperf)
    global_grad.append(gradnorms)
    global_resutls.append(results)
    global_steperf.append(emastperf)
    global_eener.append(emaener)
    global_steener.append(emastener)
```
