```python

import numpy as np
import pdb
import fire
import matplotlib.pyplot as plt
import torch
import torch.optim.swa_utils as swa_utils
from torchdiffeq import odeint_adjoint as odeint
import torchcde
import torchsde
from torch import nn
import copy
import matplotlib.pyplot as plt
from torch import profiler
from tqdm import tqdm
import numpy as np
from typing import Sequence
import pdb
from torchdiffeq import odeint
from helpers import KuramotoAGM, objective, adj_lattice, criticalK,adjoint_init, KuramotoAdjoint, grad_estimate, order_param, adabelief_update,AGM_ODE, ODE_Adjoint
import psutil
import os
import networkx as nx
from IPython.display import clear_output
is_cuda = torch.cuda.is_available()
device = 'cuda' if is_cuda else 'cpu'
#device = 'cpu'
if not is_cuda:
    print("Warning: CUDA not available; falling back to CPU but this is likely to be very slow.")
torch.cuda.empty_cache()
import gc
```

```python
batch_size = 512
scale=5
scale_kin=5

#download adjacency matrix
%store -r Astored
state_size=Astored.shape[1]

#frequencies of oscillators
freqs=torch.randn(1,state_size).to(device)/scale
freqs=freqs.repeat(batch_size, 1)
#%store -r freqs
freqss=freqs.cpu().detach().numpy()

#evaluate critical value
Kcrit, L=criticalK(Astored,state_size, scale)
Kcrit = 0.01 * Kcrit

#move tensors to device
A=torch.tensor(Astored).to(torch.float32).to(device)
L=torch.tensor(L).to(torch.float32).to(device)

#set final time
t_total=1
t_size=128
dt=torch.tensor(t_total/t_size).to(device)
ts_exp = torch.linspace(0, t_total, t_size).to(device)
# Initial conditions
ys_ran =torch.randn(batch_size, state_size).to(device)/scale_kin

# Initial energy
ys_ini = torch.zeros(batch_size, 1).to(device)
ys_exp = torch.cat((ys_ran,ys_ini),dim=1).to(device)
print(Kcrit)
print(Astored.shape)
```

```python
perf_global=[]
erpe_global=[]
Rs=[1e-3, 1e-4,1e-5, 1e-5, 1e-6]

for R in Rs:
    # control input to evaluate
    force_field_scal=torch.zeros(t_size,1,1).to(device)
    force_field=force_field_scal.repeat(1,batch_size, 1)
    # formad simulation function
    SML= AGM_ODE(batch_size, state_size, Kcrit,A, freqs, force_field, dt).to(device)
    # adabelief variables:
        #mean
    m = torch.zeros_like(force_field_scal).to(device)
        #variance: ema and max
    v = torch.zeros_like(force_field_scal).to(device)
    v_hat_max = torch.zeros_like(force_field_scal).to(device)
        #optimizer's steps
    adam_steps=0
    
    #energy penalty
  
    #learning rate
    theta=0.01
    #number of iterations
    iter = 50000
    alpha=0.95
    #list of loss and energy per epoch
    perf=[]
    erpe=[]
    
    
    #ema loss
    ema_loss=0
    #save the best performing control
    best_field=copy.deepcopy(force_field)
    score=1e10
    # adjoint evaluation funciton
    SMD= ODE_Adjoint(batch_size, state_size, Kcrit, freqs,A, force_field, ys_ran, dt, device).to(device)
    for j in tqdm(range(iter)):
        print(j)
        ys_ran =torch.randn(batch_size, state_size).to(device)/scale_kin
    
        # Initial energy
        ys_ini = torch.zeros(batch_size, 1).to(device)
        ys_exp = torch.cat((ys_ran,ys_ini),dim=1).to(device)
        #fun forward simulations
        with torch.no_grad():
            ys_tray = odeint(SML, ys_exp, ts_exp, method = 'midpoint')
    
        #calculate loss
        obj= objective(ys_tray[-1,:,0:state_size],A)
        kin_energy=R*ys_tray[-1,:,-1].mean()*state_size
        loss=obj.mean()
       
        #update ema loss
       
    
        #print current values of loss, control energy, order parameter
        print("loss")
        print(loss+kin_energy)
        print("energy")
        print(kin_energy)
        print("order param")
        print(order_param(ys_tray[-1,:,0:state_size],L))
        
        ys=ys_tray[:,:,0:state_size]
        #evaluate initial adjoint value
        initials=adjoint_init(ys[-1,:,:],state_size)
    
        #run adjoint evaluation 
        SMD.x=ys
        SMD.u=force_field
            
        with torch.no_grad():
            p_tray = odeint(SMD, initials, ts_exp, method = 'midpoint')
            
        
        #evaluate gradient
        p=torch.flip(p_tray, [0])
        grad=2*R*force_field+grad_estimate(p,ys,state_size,Kcrit)
    
        #update control signal with optimizer
        with torch.no_grad():
            force_field_scal, m, v, v_hat_max, adam_steps = adabelief_update(force_field_scal, (torch.mean(grad,dim=1)).unsqueeze(1), m, v, v_hat_max, adam_steps, lr=theta)
            
        force_field=force_field_scal.repeat(1,batch_size, 1)
        SML.u=force_field
        perf.append(loss.cpu().detach().numpy())
        erpe.append((ys_tray[-1,:,-1].mean()).cpu().detach().numpy())
        with torch.no_grad():
            torch.cuda.empty_cache()
    
        gc.collect()
        if j%1000==0:
            clear_output(wait=False)
    perf_global.append(perf)
    erpe_global.append(erpe)
    t = np.linspace(0, iter, iter)   
    plt.plot(t,np.array(perf))
    
    plt.fill_between(t, np.array(perf)-np.array(erpe), np.array(perf)+np.array(erpe))
    plt.show()
```
