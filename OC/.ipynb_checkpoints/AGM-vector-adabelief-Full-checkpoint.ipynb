```python
#more advanced version of AGM-adabelief which evaluates the signal for each component separately
import numpy as np
import pdb

import fire
import matplotlib.pyplot as plt
import torch
import torch.optim.swa_utils as swa_utils
import torchcde
import torchsde
from torch import nn
from torchdiffeq import odeint_adjoint as odeint
import copy
import matplotlib.pyplot as plt
from torch import profiler
from tqdm import tqdm
import numpy as np
from typing import Sequence
import pdb
from torchdiffeq import odeint
from helpers import KuramotoAGM, objective, adj_lattice, criticalK,adjoint_init, KuramotoAdjoint,  order_param,  adabelief_update, grad_estimate_vec, adabelief_update_manifold, combine_vector_field, AGM_ODE, ODE_Adjoint
import psutil
import os
import networkx as nx
from IPython.display import clear_output
is_cuda = torch.cuda.is_available()
device = 'cuda' if is_cuda else 'cpu'
#device = 'cpu'
if not is_cuda:
    print("Warning: CUDA not available; falling back to CPU but this is likely to be very slow.")
torch.cuda.empty_cache()
import gc
import pickle
```

```python
batch_size = 512
scale=5
scale_kin=5

%store -r Astored
state_size=Astored.shape[1]
freqs=torch.randn(1,state_size).to(device)/scale
freqs=freqs.repeat(batch_size, 1)
freqss=freqs.cpu().detach().numpy()
%store freqs

Kcrit, L=criticalK(Astored,state_size, scale)
Kcrit = 0.01 * Kcrit


A=torch.tensor(Astored).to(torch.float32).to(device)
L=torch.tensor(L).to(torch.float32).to(device)


print(Kcrit)
print(Astored.shape)
```

```python
t_total=1
t_size=64
dt=torch.tensor(t_total/t_size).to(device)
#force_field=torch.zeros(t_size,batch_size,state_size).to(device)
temporal= torch.zeros(t_size, 1).to(device)

#temporal=copy.deepcopy(best_field.squeeze(2))
components= torch.ones(1, state_size).to(device)
components= components/ torch.sqrt((torch.sum(components**2, dim=1)).unsqueeze(1))

force_field= combine_vector_field(temporal, components)

force_field=force_field.repeat(1,batch_size, 1)
SML= AGM_ODE(batch_size, state_size, Kcrit,A, freqs, force_field, dt).to(device)

    
m = torch.zeros_like(components).to(device)
v = torch.zeros_like(components).to(device)
v_hat_max = torch.zeros_like(components).to(device)

mt = torch.zeros_like(temporal).to(device)
vt = torch.zeros_like(temporal).to(device)
vt_hat_max = torch.zeros_like(temporal).to(device)


adam_steps=0
adam_steps_t=0

R=0.0001

iter = 40000

perf=[]
erpe=[]
flos=[]
t = np.linspace(0, iter, iter)

ys_ran =torch.randn(batch_size, state_size).to(device)/scale_kin

ys_ini7 = torch.zeros(batch_size, 1).to(device)
#ys_exp = torch.ones(batch_size, state_size)
ts_exp = torch.linspace(0, t_total, t_size).to(device)
ys_exp = torch.cat((ys_ran,ys_ini7),dim=1).to(device)
```

```python

sched1=1
sched2=1
alpha=1
ema_loss=0
theta=0.01
SMD= ODE_Adjoint(batch_size, state_size, Kcrit, freqs,A, force_field, ys_ran, dt, device).to(device)
for j in tqdm(range(iter)):

    ys_ran =torch.randn(batch_size, state_size).to(device)/scale_kin

    ys_ini7 = torch.zeros(batch_size, 1).to(device)
    #ys_exp = torch.ones(batch_size, state_size)
    ts_exp = torch.linspace(0, t_total, t_size).to(device)
    ys_exp = torch.cat((ys_ran,ys_ini7),dim=1).to(device)
    
    print(j)
    force_field= combine_vector_field(temporal, components)
    force_field=force_field.repeat(1,batch_size, 1)
    
    with torch.no_grad():
        ys_tray = odeint(SML, ys_exp, ts_exp, method = 'midpoint')
  
    obj= objective(ys_tray[-1,:,0:state_size],A)
    kin= R*ys_tray[-1,:,-1].mean()
    loss=obj.mean() 
  
    
  
    print("loss")
    print(loss+kin)
    print("energy")
    print(kin)

    ys=ys_tray[:,:,0:state_size]
    initials=adjoint_init(ys_tray[-1,:,0:state_size],state_size)

    SMD.x=ys
    SMD.u=force_field
    
   
    with torch.no_grad():
        p_tray = odeint(SMD, initials, ts_exp, method = 'midpoint')
    
    print("order param")
    print(order_param(ys_tray[-1,:,0:state_size],L))
    p=torch.flip(p_tray, [0])
    comp_grad, scal_grad=grad_estimate_vec(p,ys,temporal,components,state_size,R,Kcrit)
  
    beta_grad_comp=  Kcrit*torch.sum(scal_grad, dim=0)
    beta_grad_scal=  2*R*temporal +Kcrit* torch.sum(comp_grad,dim=2)
   
    with torch.no_grad():
        
        temporal, mt, vt, vt_hat_max, adam_steps_t = adabelief_update(temporal, (torch.mean(beta_grad_scal, dim=1)).unsqueeze(1), mt, vt, vt_hat_max, adam_steps_t, lr=theta*(sched1**adam_steps_t ))
    
        components, m, v, v_hat_max, adam_steps = adabelief_update_manifold(components, (torch.mean(beta_grad_comp, dim=0)).unsqueeze(0), m, v, v_hat_max, adam_steps, lr=theta*(sched2**adam_steps))
 
 
    force_field= combine_vector_field(temporal, components)
    force_field=force_field.repeat(1,batch_size, 1)
    SML.u=force_field
    perf.append(loss.cpu().detach().numpy())
    #erpe.append(stdvec.cpu().detach().numpy())
    erpe.append((ys_tray[-1,:,-1].mean()).cpu().detach().numpy())
    flos.append((loss+kin).cpu().detach().numpy())
    with torch.no_grad():
        torch.cuda.empty_cache()

    #print("energy used")
    #print( torch.mean(torch.sum(force_field[:,:,0]**2, axis=0)*dt*state_size))

    gc.collect()
    #if j%1000==0:
        #clear_output(wait=False)
plt.plot(t,np.array(perf))

plt.fill_between(t, np.array(perf)-np.array(erpe), np.array(perf)+np.array(erpe))
plt.show()
```

```python

def global_param(y, L):
    ymat=y.unsqueeze(2)
    
  
    ysin= torch.sin(y)
    ycos= torch.cos(y)

    quad_form_sin = torch.einsum('tbi,ij,tbj->tb', ysin, L, ysin)
    quad_form_cos = torch.einsum('tbi,ij,tbj->tb', ycos, L, ycos)

    quad_forms=(quad_form_sin + quad_form_cos)/(y.size(-1)**2)
    r= torch.ones_like(quad_forms)-quad_forms
    return r
ord_param=global_param(ys_tray[:,:,0:state_size],L)
```
