```python

import numpy as np
import pdb

import jax
import jax.numpy as jnp
from jax import vmap, jit
import fire
import matplotlib.pyplot as plt
import torch
import torch.optim.swa_utils as swa_utils
import torchcde
import torchsde
from torch import nn
import copy
import matplotlib.pyplot as plt
from torch import profiler
from tqdm import tqdm
import numpy as np
from typing import Sequence
import pdb
from torchdiffeq import odeint
from scipy.linalg import cholesky, inv
is_cuda = torch.cuda.is_available()
device = 'cuda' if is_cuda else 'cpu'
import psutil
import os
import sys
project_root = os.path.abspath(os.path.join(os.getcwd(), '../..')) 
sys.path.append(project_root)  # Add folder_a to sys.path
sys.path.append(os.path.join(project_root, '../..', 'Networks'))  # Add folder_b to sys.path
from Networks.utils import lqr, Kgain, batchsize_selector, lqr_batch_jax, lqr_batch_jax_heun, generate_system_canon,  grammian, random_companion_batch,  grammian_single, vt
```

```python
weight=0.1
N=128
iter=1000
nsteps=32
batch_size=2*1024
total=1
dt=total/nsteps

A,B=generate_system_canon(N)
Qfc = np.eye(N)
Q = 0 *Qfc
```

```python

batched_heun=jax.jit(lqr_batch_jax_heun, static_argnums=4)
count = 0
mean = 0.0
M2 = 0.0
for i in range(iter):

    R = weight*np.eye(B.shape[1])*dt

    amat=((np.random.rand(batch_size,N))-0.5)
    A_batch=np.broadcast_to( A, (batch_size, A.shape[0], A.shape[1]))
    B_batch= np.broadcast_to(B, (batch_size, B.shape[0], B.shape[1]))
    A_batch=np.array(A_batch)
    A_batch[:,-1,:]=amat

    A_batch=jnp.array(A_batch)
    B_batch=jnp.array(B_batch)
    
    Qfc=jnp.array(Qfc)
    Q=jnp.array(Q)
    R=jnp.array(R)
    nsteps=int(nsteps)

    optimal_control_input =batched_heun( Q, R/dt, jnp.array(A_batch), jnp.array(B_batch), nsteps, Qfc)

    x0_raw=np.random.randn(A_batch.shape[0],x0.shape[1])
    norms = np.linalg.norm(x0_raw, axis=1, keepdims=True)
    actual_state_x=x0_raw/norms

    desired_state_xf=np.broadcast_to(0*x0, (A_batch.shape[0],  x0.shape[1]))
    
    energy=0

    for l in tqdm(range(nsteps)):
       
        state_error = actual_state_x - desired_state_xf                
     
        action=np.einsum("bij, bj -> bi", optimal_control_input[:,l,:,:], state_error) 
      
        pred_state_x =  actual_state_x + dt*(np.einsum("bij, bj -> bi", A_batch, actual_state_x) + np.einsum("bij, bj -> bi", B_batch, action))

        pred_error= pred_state_x - desired_state_xf

        pred_action=np.einsum("bij, bj -> bi", optimal_control_input[:,l,:,:], pred_error) 

        actual_state_x = actual_state_x + dt*(np.einsum("bij, bj -> bi", A_batch, (actual_state_x+pred_state_x)/2) + np.einsum("bij, bj -> bi", B_batch, (action+pred_action)/2))

        energy+=  (np.einsum("bj,bj->b", action,np.einsum("ij, bj -> bi",R, (action+pred_action)/2)))
                  
        real_error=actual_state_x - desired_state_xf
        
        sem = (np.sum(real_error**2, axis=1))
 
    
    print("current mean overall loss")
    
    batch= np.array(energy+sem)
    b_count = batch.size
    b_mean = batch.mean()
    b_M2 = ((batch - b_mean) ** 2).sum()

    delta = b_mean - mean
    total = count + b_count

    mean += delta * b_count / total
    M2 += b_M2 + delta**2 * count * b_count / total

    count = total
    
   
    print(mean)
    print("current ste")
    print(np.sqrt(M2/count)/np.sqrt(count))
    
print("mean overall loss")
print(mean)
print("standard error")
print(np.sqrt(M2/count)/np.sqrt(count))
```
