```python
import gymnasium as gym
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from torch.distributions.normal import Normal
from collections import OrderedDict
import networkx as nx
import scipy.linalg
import pdb
from gymnasium import spaces
from IPython.display import clear_output
from adabelief_pytorch import AdaBelief
from kuramoto_vec import VectorizedKuramotoEnv
# Environment-specific imports and functions

# PPO Configuration
TOTAL_TIMESTEPS = 100000
STEPS_PER_TRAJECTORY =33
EPOCHS = 10
BATCH_SIZE = 2024
GAMMA = 0.999
GAE_LAMBDA = 0.95
CLIP_EPSILON = 0.2
LEARNING_RATE = 2e-4
REWARD_SCALE = 1e-5
ACTION_SCALE = 2000.0  # Kuramoto action bound [-5000, 5000]
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

```

```python
class ActorCritic(nn.Module):
    def __init__(self, obs_dim, act_dim,omniscent):
        super().__init__()
        self.omniscent=omniscent
        if omniscent:
            input_dim=obs_dim
        else:
            input_dim=obs_dim-act_dim
        mult=32   
        self.net = nn.Sequential(
            nn.Linear(input_dim, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim),
            nn.ReLU(),
        )
        
        self.mean = nn.Linear(act_dim, act_dim)

        self.radial = nn.Sequential(
            nn.Linear(1, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, 1),
        )
        self.log_std = nn.Parameter(torch.zeros(act_dim))
        self.critic = nn.Sequential(
            nn.Linear(obs_dim, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, act_dim*mult),
            nn.ReLU(),
            nn.Linear(act_dim*mult, 1),
        )
    
    def forward(self, obs):
     
        if self.omniscent:
            x = self.net(obs)
            r= self.radial(obs[act_dim:act_dim+1])
        else:
         
            if obs.dim()==1:
            
                x = self.net(obs[act_dim:])
                r= self.radial(obs[act_dim:act_dim+1])
            elif obs.dim()==2:
            
                x= self.net(obs[:,act_dim:])
                r= self.radial(obs[:,act_dim:act_dim+1])
            elif obs.dim()==3:
            
                x= self.net(obs[:,:,act_dim:])
                r= self.radial(obs[:,:,act_dim:act_dim+1])
              
        mean = self.mean(x)
    
        mean=r * mean/mean.norm(p=2, dim=-1, keepdim=True)
              
        log_std = self.log_std.expand_as(mean)
        std = torch.exp(log_std)
        value = self.critic(obs).squeeze(-1)
        return mean, std, value

def compute_gae(rewards, values, next_value, dones, gamma, lam):
    advantages = torch.zeros_like(rewards)
    returns = torch.zeros_like(rewards)
    gae = 0
    
    for t in reversed(range(len(rewards))):
        if dones[t]:  # Reset GAE if the episode ends

            gae = 0

    
        delta = rewards[t] + gamma * next_value * (1 - dones[t]) - values[t]
        gae = delta + gamma * lam * (1 - dones[t]) * gae
        advantages[t] = gae
        returns[t] = advantages[t] + values[t]
        next_value = values[t]
    
    return advantages, returns

def ppo_update(model, optimizer, obs, acts, old_log_probs, advantages, returns, clip_epsilon, batch_size):
    dataset_size = obs.shape[0]
    indices = torch.randperm(dataset_size)
    
    
   
    mean, std, values = model(obs)
    dist = Normal(mean, std)
    log_probs = dist.log_prob(acts).sum(dim=-1)

    ratios = torch.exp(log_probs - old_log_probs)
    surr1 = ratios * advantages
    surr2 = torch.clamp(ratios, 1 - clip_epsilon, 1 + clip_epsilon) * advantages
    policy_loss = -torch.min(surr1, surr2).mean()
    value_loss = ((values - returns) ** 2).mean()
    loss = policy_loss + 0.5 * value_loss
   
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()


# Generate adjacency matrix
  # Number of oscillators
%store -r Astored
A = Astored
#A = adj_lattice(DIM, "erdos", 0.3)  # Erds-Rnyi graph with p=0.5
omniscent=False

#Rs=[1e-3, 1e-4,1e-5, 1e-5, 1e-6]
Rs=[0.1]
fener=[]
floss=[]
fener_std=[]
floss_std=[]
model = ActorCritic(2*A.shape[0]+1, A.shape[0],omniscent).to(DEVICE)
for R in Rs:
# Environment setup with kwargs
    env_kwargs = {
        "A": A,
        "R": R,  # Reward penalty coefficient
        "max_t": 32,
        "noise": 0.1,
        "delayed_steps": 0
    }
    env = VectorizedKuramotoEnv(num_envs=512,**env_kwargs)
    print(env.K)
    
    obs_dim = env.observation_space.shape[-1]  # 2*dim+1 = 21
    act_dim = env.action_space.shape[-1]  # dim = 10
    
    # Initialize model and optimizer
    #model = ActorCritic(obs_dim, act_dim,omniscent).to(DEVICE)
    #optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    optimizer = AdaBelief(model.parameters(), lr=LEARNING_RATE, eps=1e-12, betas=(0.9,0.999), weight_decouple = False, rectify = False, amsgrad=True)
    
    
    # Training loop
    obs = env.reset(seed=42)
    obs = torch.tensor(obs, dtype=torch.float32, device=DEVICE)
    episode_rewards = []
    total_steps = 0
    episode_reward = 0
    
    final_energy_array=[]
    final_cost_array=[]
    final_energy_array_std=[]
    final_cost_array_std=[]
    while total_steps < TOTAL_TIMESTEPS:
        rollout_obs = []
        rollout_acts = []
        rollout_log_probs = []
        rollout_rewards = []
        rollout_values = []
        rollout_dones = []
        total_energy=0
        energy_array=[]
        cost_array=[]
        orparam_array=[]
        # Collect trajectory
        for _ in range(STEPS_PER_TRAJECTORY):
            with torch.no_grad():
             
                mean, std, value = model(obs)
                dist = Normal(mean, std)
                action = dist.sample()
                log_prob = dist.log_prob(action).sum(dim=-1)
                
                # Scale action to [-5000, 5000]
                action_scaled = action
            
            rollout_obs.append(obs.clone())
            rollout_acts.append(action)
            rollout_log_probs.append(log_prob)
            rollout_values.append(value)
            
            # Step environment
            next_obs, reward, terminated, truncated,  final_costs, energies, order_params = env.step(action_scaled.cpu().numpy())
        
            done = terminated[0] or truncated[0]
            total_energy+=energies
            
           
            rollout_rewards.append(torch.tensor(reward, dtype=torch.float32, device=DEVICE))
            rollout_dones.append(torch.tensor(done, dtype=torch.float32, device=DEVICE))
            
            obs = torch.tensor(next_obs, dtype=torch.float32, device=DEVICE)
            
            episode_reward += reward
            total_steps += 1
           
            if done:
                
                episode_rewards.append(episode_reward)
                episode_reward = 0
                obs= env.reset()
                obs = torch.tensor(obs, dtype=torch.float32, device=DEVICE)
                energy_array.append(total_energy)
                cost_array.append(final_costs)
                orparam_array.append(order_params)
                #print("total energy")
                #print(total_energy)
                total_energy=0
                break
                #print("final cost")
                #print(final_cost)
    
        print("average energy")
        print(np.mean(energy_array[0]))
        print("average cost")
        print(np.mean(cost_array[0]))
        print("average order param")
        print(np.mean(orparam_array[0]))
    
        final_energy_array.append(np.mean(energy_array[0]))
        final_cost_array.append(np.mean(cost_array[0]))    
        final_energy_array_std.append(np.std(energy_array[0])/np.sqrt(len(energy_array[0])))
        final_cost_array_std.append(np.std(cost_array[0])/np.sqrt(len(cost_array[0])))   
        # Process rollout

        rollout_obs = torch.stack(rollout_obs)
        rollout_acts = torch.stack(rollout_acts)
        rollout_log_probs = torch.stack(rollout_log_probs)
        rollout_rewards = torch.stack(rollout_rewards)
        rollout_values = torch.stack(rollout_values)
        rollout_dones = torch.stack(rollout_dones)
       
        # Compute next value for GAE
        with torch.no_grad():
            _, _, next_value = model(obs)
        
        # Compute advantages and returns
        advantages, returns = compute_gae(
            rollout_rewards, rollout_values, next_value, rollout_dones, GAMMA, GAE_LAMBDA
        )
       
        # Normalize advantages
        #advantages = (advantages - advantages.mean()) / (advantages.std() + 1e-8)
        
    
        
        ppo_update(
            model, optimizer, rollout_obs, rollout_acts, rollout_log_probs,
            advantages, returns, CLIP_EPSILON, BATCH_SIZE
        )
         # Print progress
        if episode_rewards:
            print(f"Steps: {total_steps}, Avg Reward: {np.mean(episode_rewards[-100:]):.4f}")

        del rollout_obs, rollout_acts, rollout_log_probs, rollout_rewards, rollout_values, rollout_dones, advantages, returns
    clear_output(wait=False)
    fener.append(final_energy_array)
    floss.append(final_cost_array)
    fener_std.append(final_energy_array_std)
    floss_std.append(final_cost_array_std)
    env.close()
```
